## **1. ლოლიპოპის დიაგრამა (Lollipop Chart) [1 ქულა]:**

მოცემულია მონაცემები საქართველოს რამდენიმე ქალაქში საზოგადოებრივი ბაღების რაოდენობის შესახებ (იხ. კოდი)

ლოლიპოპის დიაგრამით ვიზუალურად აჩვენეთ ბაღების განაწილება ქალაქების მიხედვით. დაალაგეთ გრაფიკი კლებადობით და დაამატეთ სათაური.


```python
import matplotlib.pyplot as plt
import random

cities = ['თბილისი', 'ქუთაისი', 'ბათუმი', 'რუსთავი', 'გორი', 'ზუგდიდი', 'ფოთი', 'ხაშური', 'სამტრედია', 'სენაკი']
kindergartens = [random.randint(50, 200) for _ in range(len(cities))]

# მონაცემების დალაგება კლებადობით
sorted_data = sorted(zip(kindergartens, cities), reverse=True)
sorted_kindergartens, sorted_cities = zip(*sorted_data)

# ლოლიპოპის დიაგრამის შექმნა
plt.figure(figsize=(10, 6))
markerline, stemlines, baseline = plt.stem(sorted_cities, sorted_kindergartens, basefmt=" ")
plt.setp(stemlines, linestyle="--", color="blue", linewidth=1)
plt.setp(markerline, color="red", markersize=8)

# გრაფიკის დიზაინი
plt.title("ბაღების რაოდენობა საქართველოს ქალაქებში", fontsize=16)
plt.xlabel("ქალაქები", fontsize=12)
plt.ylabel("ბაღების რაოდენობა", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()

# გრაფიკის ჩვენება
plt.show()

```

## **2. რადარული დიაგრამა (Radar/Spider Chart) [1 ქულა]:**

კომპანია აფასებს 3 პროფესიის (მონაცემთა მეცნიერი, მონაცემთა ინჟინერი, მონაცემთა ანალიტიკოსი) უნარებს 1-დან 10-მდე ქულით. მონაცემები მოცემულია კოდში.

ააგეთ რადარული დიაგრამა, რომელიც ერთდროულად აჩვენებს ყველა პროფილის ყველა უნარის შეფასებას. დიაგრამა უნდა იყოს ვიზუალურად მკაფიო.


```python
import matplotlib.pyplot as plt
import numpy as np

# პროფესიები
roles = ['მონაცემთა მეცნიერი', 'მონაცემთა ინჟინერი', 'მონაცემთა ანალიტიკოსი']

# უნარები
criteria = [
    'Python', 'SQL', 'Machine Learning', 'Data Visualization',
    'Big Data', 'Cloud Platforms', 'Statistics', 'Communication Skills', 'ETL Pipelines'
]

# უნარების ქულები
scores = [
    [9, 8, 10, 7, 7, 8, 9, 6, 5],  # მონაცემთა მეცნიერი
    [8, 9, 6, 6, 9, 10, 7, 5, 10],  # მონაცემთა ინჟინერი
    [7, 10, 5, 9, 6, 7, 8, 8, 6]   # მონაცემთა ანალიტიკოსი
]

# რადარის დიაგრამის შექმნისთვის მომზადება
num_criteria = len(criteria)
angles = np.linspace(0, 2 * np.pi, num_criteria, endpoint=False).tolist()
angles += angles[:1]  # დახურვა

# მონაცემების დახურვა რადარის დიაგრამისთვის
scores = [score + score[:1] for score in scores]

# გრაფიკის შექმნა
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection': 'polar'})

# თითოეული პროფილის დამატება
for role, score in zip(roles, scores):
    ax.plot(angles, score, linewidth=2, label=role)
    ax.fill(angles, score, alpha=0.25)

# უნარების სახელების დამატება
ax.set_xticks(angles[:-1])
ax.set_xticklabels(criteria, fontsize=10)

# რადარის დიაგრამის დიზაინი
ax.set_rlabel_position(0)
ax.yaxis.grid(True, linestyle="--", alpha=0.7)
ax.xaxis.grid(True, linestyle="--", alpha=0.7)
ax.set_yticks([2, 4, 6, 8, 10])
ax.set_yticklabels(['2', '4', '6', '8', '10'], fontsize=8)
plt.title("უნარების შეფასება პროფესიების მიხედვით", fontsize=16, pad=20)

# ლეგენდა
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# გრაფიკის ჩვენება
plt.tight_layout()
plt.show()
```

## **3. სანკის დიაგრამა (Sankey Diagram) [1 ქულა]:**

კომპანია იკვლევს მომხმარებელთა ქცევას ვებ-გვერდზე. მოცემულია მონაცემები იმის შესახებ, თუ რამდენმა ვიზიტორმა მოინახულა სხვადასხვა გვერდები და საბოლოოდ რამდენმა იყიდა პროდუქტი (იხ. კოდი).

ააგეთ სანკის დიაგრამა, რომელიც აჩვენებს მომხმარებელთა ნაკადს ვებ-გვერდის გვერდებს შორის. ფერებით გამოყავით ვებ-გვერდის თითოეული გვერდი და ნაკადებზე მიუთითეთ ვიზიტორთა რაოდენობა.


```python
import plotly.graph_objects as go
import random

main_page = random.randint(400, 600)
product_page = random.randint(200, 400)
cart_page = random.randint(100, 300)
purchased = random.randint(50, 150)

labels = ["Main Page", "Product Page", "Cart Page", "Purchased"]
sources = [0, 1, 2]
targets = [1, 2, 3]
values = [main_page, product_page, cart_page]

fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=labels,
        color=["blue", "green", "orange", "red"]
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values
    )
))

fig.update_layout(title_text="User Flow Through Website", font_size=10)
fig.show()
```

## **4. ვიოლინოს დიაგრამა (Violin Plot) [1 ქულა]:**

IT კომპანია სხვადასხვა პოზიციებზე ხელფასების განაწილებას ადარებს.

შექმენით ვიოლინოს დიაგრამა, რომელიც ერთდროულად აჩვენებს როგორც ხელფასების განაწილებას, ასევე კვარტილების საზღვრებს თითოეული პოზიციისთვის. გამოიყენეთ Seaborn ბიბლიოთეკა. დაამატეთ შესაბამისი ფერები და მონიშვნები.




```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# მონაცემთა გენერირება
positions = ['Junior Developer', 'Middle Developer', 'Senior Developer', 'Team Lead', 'Project Manager']
salaries = {pos: list(np.random.normal(loc=1500 + i * 1000, scale=200, size=30)) for i, pos in enumerate(positions)}

df = pd.DataFrame([(pos, salary) for pos in positions for salary in salaries[pos]], columns=['Position', 'Salary'])

# ვიოლინოს დიაგრამის შექმნა
plt.figure(figsize=(10, 6))
sns.violinplot(x='Position', y='Salary', data=df, palette='Set2', inner='quartile')

# გრაფიკის სათაური და მონიშვნები
plt.title('ხელფასების განაწილება სხვადასხვა პოზიციებზე', fontsize=16)
plt.xlabel('პოზიცია', fontsize=12)
plt.ylabel('ხელფასი (ლარი)', fontsize=12)
plt.xticks(rotation=15, fontsize=10)
plt.yticks(fontsize=10)

# გრაფიკის ჩვენება
plt.tight_layout()
plt.show()
```

## **5. კორელაციის მატრიცა სითბური რუკით (Correlation Matrix Heatmap) [1 ქულა]:**

მოცემულია 5 სხვადასხვა პორტფელის ისტორიული მონაცემები (იხ. კოდი)

გამოთვალეთ პორტფელებს შორის კორელაციის კოეფიციენტები და ააგეთ სიმეტრიული სითბური რუკა (heatmap) კორელაციების მატრიცის საფუძველზე. გამოიყენეთ Seaborn ბიბლიოთეკა შესაბამისი ფერების სქემით და გრადაციით. დაამატეთ ფერთა სკალა და  მონიშვნები საჭიროებისამებრ.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import random
import pandas as pd

portfolio_values = [
    [random.uniform(1000, 1500) for _ in range(12)],
    [random.uniform(800, 1200) for _ in range(12)],
    [random.uniform(1200, 1800) for _ in range(12)],
    [random.uniform(900, 1300) for _ in range(12)],
    [random.uniform(1400, 2000) for _ in range(12)]
]

df = pd.DataFrame(portfolio_values).T
df.columns = [f'Portfolio {i+1}' for i in range(len(portfolio_values))]

# კორელაციის მატრიცის გამოთვლა
correlation_matrix = df.corr()

# სითბური რუკის აგება
plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix,
    annot=True,  # ანოტაციის დამატება (მნიშვნელობების ჩვენება)
    cmap="coolwarm",  # ფერთა სქემა
    fmt=".2f",  # მნიშვნელობების ფორმატი
    linewidths=0.5,  # ხაზების სიგანე
    cbar_kws={'label': 'Correlation Coefficient'}  # ფერთა სკალას სათაური
)

# გრაფიკის სათაური და მონიშვნები
plt.title('პორტფელების კორელაციის სითბური რუკა', fontsize=16)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)

# გრაფიკის ჩვენება
plt.tight_layout()
plt.show()
```

## **6. პარალელური კოორდინატების დიაგრამა (Parallel Coordinates Plot) [1 ქულა]:**

ავიაკომპანიას აქვს მგზავრების შესახებ მონაცემები. ქვემოთ მოცემულია სამი მგზავრის მონაცემები მათი ქცევის ისტორიის შესახებ (იხ. კოდი).

ააგეთ პარალელური კოორდინატების დიაგრამა, სადაც x ღერძზე იქნება თითოეული კატეგორია, ხოლო y ღერძზე მათი შესაბამისი მნიშვნელობები. თითოეული მგზავრი წარმოდგენილი იყოს ტეხილით. გამოიყენეთ Plotly Express ან Plotly Graph Objects მოდულები.



```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import random

random.seed(22)

# მგზავრების სია
passengers = ['Passenger 1', 'Passenger 2', 'Passenger 3', 'Passenger 4', 'Passenger 5']

# მონაცემთა გენერირება
data = {
    'Trips': [random.randint(5, 15) for _ in range(5)],  # მოგზაურობების რაოდენობა
}
data['Flights'] = [trips * random.randint(1, 3) for trips in data['Trips']]
data['Duration'] = [flights * random.randint(2, 6) for flights in data['Flights']]
data['Kilometers'] = [duration * random.randint(100, 600) for duration in data['Duration']]
data['Reward'] = [Kilometers // 10 for Kilometers in data['Kilometers']]

# მონაცემების ჩაწერა DataFrame-ში
df = pd.DataFrame(data, index=passengers)

# DataFrame-ის აღდგენა Plotly-სთვის
df.reset_index(inplace=True)
df.rename(columns={'index': 'Passenger'}, inplace=True)

# პარალელური კოორდინატების დიაგრამის შექმნა
fig = px.parallel_coordinates(
    df,
    dimensions=['Trips', 'Flights', 'Duration', 'Kilometers', 'Reward'],
    color='Reward',  # ფერის დაშვება "Reward" სვეტზე
    color_continuous_scale=px.colors.diverging.Tealrose,
    labels={'Trips': 'მოგზაურობები', 'Flights': 'რეისები', 'Duration': 'ხანგრძლივობა',
            'Kilometers': 'კილომეტრები', 'Reward': 'ჯილდო'},
)

# სათაურის დამატება და დიაგრამის ჩვენება
fig.update_layout(title='მგზავრების ქცევის პარალელური კოორდინატები')
fig.show()
```

## **7. თემატური ჯგუფების წილის ვიზუალიზაცია (Treemap) [1 ქულა]:**

მომხმარებლები მონაწილეობენ 5 თემატურ ჯგუფში: ტექნოლოგიები, სპორტი, მოგზაურობა, ჯანმრთელობა, კულტურა.  
თქვენ გაქვთ მონაცემები, რომლებიც ასახავს თითოეულ თემატურ ჯგუფში მომხმარებელთა აქტივობების რაოდენობას (იხ. კოდი).

1. ააგეთ Treemap, რომელიც ასახავს თითოეული თემატური ჯგუფის წილს საერთო აქტივობებში.  
2. გრაფიკზე უნდა ჩანდეს:  
   - ჯგუფის სახელი.  
   - აქტივობების რაოდენობა.  
   - ფერები, რაც თემებს განასხვავებს.

გამოიყენეთ squarify

```python
import matplotlib.pyplot as plt
import squarify
import random

# თემები და აქტივობების რაოდენობა
groups = ['ტექნოლოგიები', 'სპორტი', 'მოგზაურობა', 'ჯანმრთელობა', 'კულტურა']
activities = [random.randint(50, 200) for _ in groups]

# საერთო აქტივობების გამოთვლა
total_activities = sum(activities)

# ფერების გენერირება
colors = ['#FF5733', '#33FF57', '#3357FF', '#FF33A6', '#A633FF']

# Treemap მონაცემების ვიზუალიზაცია
plt.figure(figsize=(10, 6))
squarify.plot(
    sizes=activities,
    label=[f"{group}\n{activity}" for group, activity in zip(groups, activities)],
    color=colors,
    alpha=0.8
)

# სათაური და დიზაინის დამატება
plt.title('თემატური ჯგუფების წილის ვიზუალიზაცია')
plt.axis('off')  # ღერძების გამორთვა
plt.show()
```
